```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("HPS_default.csv")
print(list(df))

df.plot(kind='scatter', x='Best Fitness', y='Average Fitness')
plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_trisurf(df["Crossover probability"], 
                df["Mutation probability"], 
                df["Best Fitness"], 
                cmap=cm.jet)
ax.set_xlabel("Crossover probability")
ax.set_ylabel("Mutation probability")
ax.set_zlabel("Best Fitness")
```

```python
%matplotlib notebook
import numpy as np

xlabel="Number of generations"
ylabel="Best Fitness"
fn = np.poly1d(np.polyfit(df[xlabel], df[ylabel], 1))
x=df[xlabel]
y=df[ylabel]
plt.plot(x, y, 'o', x, fn(x))
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.show()
```

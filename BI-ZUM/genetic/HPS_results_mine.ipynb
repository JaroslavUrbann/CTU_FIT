```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("HPS_mine.csv")
print(list(df))

df.plot(kind='scatter', x='Best fitness', y='Average fitness')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_trisurf(df["Tournament size"], 
                df["Tournament winners"], 
                df["Best fitness"], 
                cmap=cm.jet)
ax.set_xlabel("Tournament size")
ax.set_ylabel("Tournament winners")
ax.set_zlabel("Best fitness")
```

```python
import numpy as np
def plt2d(xlabel, ylabel="Best fitness"):
    fig, axes = plt.subplots()
    fn = np.poly1d(np.polyfit(df[xlabel], df[ylabel], 1))
    x=df[xlabel]
    y=df[ylabel]
    axes.plot(x, y, 'o', x, fn(x))
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
plt2d("Mutation probability")
plt2d("End probability")
plt2d("Nodes to add")
plt2d("Number of generations")
plt2d("Population size")
```

```python
plt2d("Population size", "Time")
plt2d("Number of generations", "Time")
plt2d("End probability", "Time")
```

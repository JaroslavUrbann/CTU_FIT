```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("default_algo.csv")
print(list(df))

df.plot(kind='scatter', x='Best Fitness', y='Average Fitness')
plt.show()
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_trisurf(df["Crossover probability"], 
                df["Mutation probability"], 
                df["Best Fitness"], 
                cmap=cm.jet)
ax.set_xlabel("Crossover probability")
ax.set_ylabel("Mutation probability")
ax.set_zlabel("Best Fitness")
```

```python
%matplotlib notebook
import numpy as np

# with early stopping of 900!!

fn = np.poly1d(np.polyfit(df["Number of generations"], df["Best Fitness"], 1))
x=df['Number of generations']
y=df['Best Fitness']
plt.plot(x, y, 'o', x, fn(x))
plt.xlabel("Number of generations")
plt.ylabel("Best Fitness")
plt.show()
```
